```python
def w_evolve(W, x, alpha = 0.01):

    dW = alpha * (np.identity(W.shape[0]) -np.outer(x ,x.T)) 
    return dW

n = 5  # Size of the matrix
W = np.random.rand(n, n)  # Generate a random matrix

activities = np.random.rand(n)

timesteps = 1000000

alpha = 0.01
dt = 0.001

values = []
for i in range(timesteps):
    # simulate it like a diffeq dumby
    new_timestep = activities + (dt * np.dot(W, activities))
    W += w_evolve(W, activities)
    values.append(new_timestep)
    activities = new_timestep

values = np.vstack(values)
print(W.shape)
```

```python
import matplotlib.pyplot as plt

for column in values.T:
    plt.plot(column)

plt.xlabel('Timestep')
plt.ylabel('Value')
plt.title('Line Plot for Each Column in Values')
plt.show()
```
